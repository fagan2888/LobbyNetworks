## Openning the dataÂ¶

```python
#function that open the file and create a "good" dataset
def createDataFrame(file):
    ##lists of 50 data are loaded and then merged
    #create list of dataframes and open file
    dataframes=[]
    file=open(dataDir+file, 'r')
    
    #counter is important for the index
    c=0
    while True:
        #each line is 50 data (as limit=50 when downloading)
        txt=file.readline()
        #possibly stop
        if txt=='':
            break

        #load the data with right index
        j=json.loads(txt)
        l=len(j)
        dataframes.append(pd.DataFrame(j, index=list(range(c, c+l))))
        
        c+=l
    #merge all datasets and return result
    return pd.concat(dataframes)

```

### Selecting events from events Data Frame:Â¶

```python
import re

def get_sum(s):
    if s is None:
        return None
    else:
        l=re.findall(r'[\d.]+', s.replace(",", ""))
        s=0
        for i in l:
            try:
                s+=float(i)
            except:
                pass
        return s
        #return sum([float(i) for i in l])

print(get_sum(" PAC $1000; Individual $250"))
print(get_sum("Sponsor $2,500;PAC $1,000;\nIndividual $500;"))
print(get_sum("Raised 208.68;"))
```

```python
dfEvent['contributions_info']=dfEvent['contributions_info'].apply(get_sum)
```

# Try to plot as a network

```python
import networkx as nx
```

```python
## maximum number of nodes:
nb=25
```

```python
dfB=dfEvents2['beneficiaries_id'][:nb]
dfH=dfEvents2['hosts_id'][:nb]
print(dfB.unique()[0])

print(len(dfB.unique()), len(dfH.unique()))

fB={}
for i in range(len(dfB.unique())):
    fB[str(dfB.unique()[i])]=i
fH={}
for i in range(len(dfH.unique())):
    fH[str(dfH.unique()[i])]=i

    
    
############################################
print('\n')
############################################
    
    
G = nx.Graph()

for i in range(len(dfB.unique())):
    G.add_node("b"+str(i))
for i in range(len(dfH.unique())):
    G.add_node("h"+str(i))

    
#G.add_nodes_from(range(len(dfB.unique()) + len(dfH.unique())))
for i in range(len(dfB)):
    #G.add_nodes_from()
    s1='h'+str(fH[str(dfH.loc[i])])#+len(dfB.unique())
    s2='b'+str(fB[str(dfB.loc[i])])
    G.add_edge(s1, s2)

print(len(dfB.unique()))
print(G.number_of_nodes())
print(G.number_of_edges())




############################################




#color stuff
c1=[-1 for i in range(len(dfB.unique()))]
c2=[1 for i in range(len(dfH.unique()))]
color=c1+c2
a=G.nodes()
a.sort()
```

```python
#execute this cell more then once to get better drawings (there is some random involved during the drawing)
plt.subplot()#figsize(8,6))
nx.draw(G, with_labels=True, font_weight='bold', node_color=color, vmin=-2,vmax=2, nodelist=a, cmap='autumn')
plt.show()
```

```python
#second type of graph
nx.draw_shell(G, with_labels=True, font_weight='bold', node_color=color, vmin=-2,vmax=2, nodelist=a, cmap='autumn')
plt.show()
```
